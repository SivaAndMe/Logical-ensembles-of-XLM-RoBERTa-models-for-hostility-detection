```
import pandas as pd
import numpy as np

df_t = pd.read_csv('/content/drive/MyDrive/constraint2021/data/test_cleaned.csv')
```

```
X_t=df_t['text']
```

```
from flair.embeddings import FlairEmbeddings, TransformerWordEmbeddings
from flair.data import Sentence
xlmr_embedding = TransformerWordEmbeddings('xlm-roberta-large', layers='-1')
# init multilingual BERT
bert_embedding = TransformerWordEmbeddings('bert-base-multilingual-uncased',layers='-1')
```

```
from flair.embeddings import StackedEmbeddings

# now create the StackedEmbedding object that combines all embeddings
stacked_embeddings = StackedEmbeddings(
    embeddings=[xlmr_embedding, bert_embedding])
```

```
import torch
```

```
def mean(sent,i):
  v_t=torch.zeros([1,1792],device='cuda:0')
  for j in range(len(sent)):
      try:
        v_t = v_t+sent[j].embedding
      except Exception as e:
        print(e)
        print(v_t.shape,sent[j].embedding.shape)
  v_t/=len(sent)
  if(i%100==0):
    print(i)
  return v_t
#started at 11.25 am ended at 11.30
```

```
from flair.data import Sentence
# sents = [Sentence(X) for X in X_t]
print("HI")
sents = [Sentence(X) for X in X_t]#[:100]
stacked_embeddings.embed(sents)
embeds = [mean(sents[i],i) for i in range(len(sents))]
```

```
embeds = [x.cpu().numpy() for x in embeds]
print(len(embeds))
embeds[0].shape
```
